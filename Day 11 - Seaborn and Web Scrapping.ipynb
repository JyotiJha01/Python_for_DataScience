# Top 10 Super Categories by Sales

```python
df['Super Category'].nunique()
```

```python
df.groupby('Super Category')['Sales (shipping price included)'].sum()
```

```python
top10scs = df.groupby('Super Category', as_index = False)['Sales (shipping price included)'].sum().sort_values(
    by = "Sales (shipping price included)", ascending = False)[:10].reset_index(drop = True)
top10scs['Sales (in $M)'] = top10scs['Sales (shipping price included)']/1000000
top10scs = top10scs.drop(columns = 'Sales (shipping price included)')
top10scs
```

```python
import requests as re
from bs4 import BeautifulSoup as bs
```

```python
url = "https://en.wikipedia.org/wiki/World_War_II"
data = re.get(url)
parser = bs(data.text, "html")
print(parser.find("title"))
```

```python
# All the data from the above webpage
raw_data = parser.find_all("table")
for i in raw_data:
    print(i.text)
```

```python
url = "https://en.wikipedia.org/wiki/Foreign_exchange_market"
data = re.get(url)
parser = bs(data.text, "html")
print(parser.find("title"))
```

```python
# All the data from the above webpage
raw_data = parser.find_all(id = "mw-panel-toc")
for i in raw_data:
    print(i.text)
```
